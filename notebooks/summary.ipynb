```python
fig, axes = plt.subplots(figsize=(12,4))

weeks = range(1,14)

max_val = {}
for function in range(1,9):

    for week in weeks:
        PROJECT_DIR, data_folder_name, week_folder_name = data_loading.get_working_week(week, function)
        df = data_loading.load_initial_data(PROJECT_DIR, data_folder_name)
        df = data_loading.apply_updates(df, week, data_folder_name)

        max_val[week] = df["Y1"].max()

    x = list(max_val.keys())
    y = list(max_val.values())

    max_ = np.max(y)
    min_ = np.min(y)

    y_scaled = (y - min_)/(max_-min_)

    plt.plot(x, y_scaled,marker='o',linewidth=2, label=function)

plt.axvspan(1, 10.5, color="royalblue", alpha=0.25)
plt.axvspan(10.5, 13, color="red", alpha=0.25)
plt.ylim(-0.1,1.2)
plt.xlim(1,13)
plt.xticks(range(1,14))
plt.yticks(range(0,2))
plt.xlabel("week")
plt.ylabel("$y/y_{max}$")
plt.legend(title="Function")
plt.title("Week-by-week increase in observed function maximum")

plt.annotate(
    "Phase 1",
    xy=(5, 1.1),          # point to annotate
    xytext=(5, 1.1),   # location of the text
    fontsize=16,
    color = 'royalblue'
)

plt.annotate(
    "Phase 2",
    xy=(11.5, 1.1),          # point to annotate
    xytext=(11.5, 1.1),   # location of the text
    fontsize=16,
    color = 'red'
)

plt.savefig("summary.png", dpi=300, bbox_inches="tight")
```
